### **Control-FLow**

```python
#### If else statement
money = int(input("Please Provide me the Money:- "))

if money == 200:
    print("I will buy a Burger")
elif money == 300:
    print("I will buy a Pizza")
elif money == 500:
    print("I will buy a Chicken Biryani")
else:
    print("I will eat at home")
```

### **Accept Two numbers and print Greater Number**

```python
num1 =     int(input("Please Provide me the First Number:- "))
num2 =    int(input("Please Provide me the Second Number:- "))
if num1 > num2:
    print(f"{num1} is Greater than {num2}")
else:
    print(f"{num2} is Greater than {num1}")
    
```
